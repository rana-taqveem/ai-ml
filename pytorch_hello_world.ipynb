```python
# Tensor Math and Ooperations


x = torch.tensor([1, 2, 3])
y = torch.tensor([8, 7, 9])

z = torch.empty(3)

z = torch.add(x, y, out=z)
print(z)
z = torch.add(x, y)
print(z)
z =  x + y
print(z)

z = torch.subtract(x, y)
print(z)
z = torch.sub(x,y)
print(z)
z = x-y
print(z)

print(torch.div(x,y))
print(torch.true_divide(x,y))

print(x>0)
print(x<0)

print(x ** 2)

M1 = torch.rand(5,3)
M2 = torch.rand(3,5)
M3 = torch.mm(M1, M2)
print(M3)

print(torch.matrix_exp(M3))

batch = 32
X1 = torch.rand((batch, 10, 12))
X2 = torch.rand((batch, 12, 10))
X3 = torch.bmm(X1, X2)
print(X3.shape)
```
